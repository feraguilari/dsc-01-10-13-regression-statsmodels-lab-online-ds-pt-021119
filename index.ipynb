Which advertising channel has a strong relationship with sales volume, and can be used to model and predict the sales. 

### Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
# Check the columns and first few rows
df = pd.read_csv("Advertising.csv", index_col=0)
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset
#The dataset contains budget spent in 3 different categories: TV, radio, and newspaper per period.
#It also contains total sales made per period.
#Most of the budget goes into TV advertising followed by newspaper and funally by radio.
```

### Step 2: Plot histograms with kde overlay to check for the normality of the predictors

```python
# For all the variables, check if they hold normality assumption
for col in df:
    df[col].plot.hist(density=True, label ='histogram')
    df[col].plot.kde(label ='kde')
    plt.title(col+' Histogram and KDE')
    plt.legend()
    plt.show()
```

```python
# Record your observations on normality here 
#Both tv and radio are somewhat simmetrical.
#radio has a very large positive skew.
#sales is slightly positively skewed.
```

### Step 3: Test for the linearity assumption. 
Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(18, 6))
for idx, channel in enumerate(['TV', 'radio', 'newspaper']):
    df.plot(kind='scatter', x=channel, y='sales', ax=axs[idx], label=channel)
plt.legend()
plt.show()
```

```python
# Record yor observations on linearity here 
#The most linear relationship is dislplaye by TV.
#Radio is not as good because there is too much variance.
#newspaper is the worse since there is no apparent linear relationship due to great variance.
```

### Conclusion so far !

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We shall move ahead with our analysis using TV and radio , and count out the newspaper due to the fact that data violates ols assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. We shall talk about this in the next section. 

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
import statsmodels.formula.api as smf
# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = smf.ols(formula=f, data=df).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"

R-suqared is .612 memaning that 61.2% of the variation in sales can be explained using the budget on TV.
For each unit change of TV budget we have  .0475 change in unit of sales.


Note here that the coefficients represent associations, not causations


### Step 6:  Draw a prediction line with data points omn a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
TV_minmax = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
print(TV_minmax.head())

# make predictions for those x values and store them
predict = model.predict(TV_minmax)
print (predict)

# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='TV', y='sales', label='TV data')
plt.plot(TV_minmax, predict, color='red', label='predictions')
plt.legend()
plt.show()
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on residuals
#Heteroskedasticity is noticed in the two upper plots since the dots seem to be diverging with higher values of x.
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
# build the formula 
g = 'sales~radio'
# create a fitted model in one line
model_radio = smf.ols(formula=g, data=df).fit()
radio_minmax = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
print(radio_minmax.head())

# make predictions for those x values and store them
predict = model.predict(radio_minmax)
print (predict)

# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='radio', y='sales', label='radio data')
plt.plot(radio_minmax, predict, color='red', label='predictions')
plt.legend()
plt.show()
```

```python
model_radio.summary()
```

```python
# Record your observations here for goodnes of fit 
#R-suqared is .332 memaning that 33.2% of the variation in sales can be explained using the budget on TV.
#For each unit change of radio budget we have  .2025 change in unit of sales.
```

```python
radio_fig = plt.figure(figsize=(15,8))
radio_fig = sm.graphics.plot_regress_exog(model_radio, "radio", fig=radio_fig)
plt.show()
```

As with TV, there is heteroskadicity.
Radio is a worse predictor of sales given its lower value of r-squared.
